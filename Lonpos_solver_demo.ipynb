```python
!pip install git+https://github.com/geraschenko/lonpos_solver
import lonpos_solver
import matplotlib.pyplot as plt
import numpy as np
```

```python
#@title Find a couple of solutions on each 2D board type. {vertical-output: true}
NUM = 2

solved_boards = []
for board_type in ['rectangle', 'arrowhead', 'triangle', 'butterfly']:
  game = lonpos_solver.Lonpos2D(board_type)
  for board, i in zip(game.solutions(), range(NUM)):
    solved_boards.append(board)
    game.plot(board)
    plt.show()
```

```python
#@title Remove a few pieces from a position and find all ways to solve from there. {vertical-output: true}

game.set_board(solved_boards[0])
game.unplace('A', 'B', 'C', 'E')

print('Given position:')
game.plot()
plt.show()

print('All solutions:')
for b in game.solutions():
  game.plot(b)
  plt.show()
```

```python
#@title Find all ways to solve a 3D position. {vertical-output: true}
#@markdown Note: finding solutions from scratch works for the 3D board, but takes much longer (minutes!). We're very important people who don't have that kind of time to blow on this.

game = lonpos_solver.Lonpos3D()
game.place('C', np.array(((0, 3, 0), (0, 4, 0), (1, 4, 0), (2, 4, 0), (3, 4, 0))))
game.place('E', np.array(((0, 2, 0), (1, 2, 0), (1, 3, 0), (2, 3, 0), (3, 3, 0))))
game.place('I', np.array(((0, 0, 0), (0, 1, 0), (1, 0, 0), (2, 0, 0), (2, 1, 0))))
game.place('F', np.array(((3, 2, 0), (4, 2, 0), (4, 3, 0))))
game.place('K', np.array(((3, 0, 0), (3, 1, 0), (4, 0, 0), (4, 1, 0))))

# I happen to know there are two solutions here.
fig, ax = plt.subplots(1, 3, figsize=(8, 6))
print('Given position and all solutions:')

game.plot(ax=ax[0])
i = 1
for b in game.solutions():
  game.plot(b, ax=ax[i])
  i += 1
```

```python
#@title Find solutions for the [puzzle-a-day calendar](https://www.dragonfjord.com/product/a-puzzle-a-day). {vertical-output: true}
game = lonpos_solver.Calendar('Jun', 15)
solutions = list(game.solutions())
print(f'{len(solutions)} solutions found for {game}. Here\'s one:')
game.plot(solutions[0])
```
